```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Concentration
t = 1
A = 0.01
M = 1
def f(x, y):
    return (M/(4*(np.pi)*A*t))*np.exp(-(((x-t)**2)+((y-t)**2))/4*A*t)
```

```python
x = np.linspace(-3, 7, 50)
y = np.linspace(-3, 7, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)


plt.contourf(X, Y, Z, 20, cmap='jet')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```
