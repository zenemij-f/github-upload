# The Eady Problem

Let's study the a particular case of Baroclinic Instability. In this configuration, we make the following assumptions:

1. The motion takes place in the *f*-plane
2. The fluid is uniformly stratified $\longrightarrow N^2$ is constant
3. The basic state zonal flow has uniform shear $\longrightarrow U(z) = U\frac{z}{H}$, where *U* is the zonal velocity at z  H.
4. The motion is constrained by two rigid and flat surfaces.

### Case 1: Unstable waves

```python
import matplotlib.pyplot as plt
import numpy as np


#Functions:

def phi(z):
    return np.cosh(mu*z/1000) - (U/mu*c)*np.sinh(mu*z/1000)

def psi(x,z):
    return (phi(z)*np.sin(l*y)*np.exp(1j*k*(x/1000-c*t))).real

def phi_prime(z):
    return mu*np.sinh(mu*z/1000) - (U/c)*np.cosh(mu*z/1000)

def temperature(x,z):
    return (phi_prime(z)*np.sin(l*y)*np.exp(1j*k*(x/1000-c*t))).real

def meridional_velocity(x,z):
    return (phi(z)*np.sin(l*y)*np.exp(1j*k*(x/1000-c*t))*1j*k).real
```

```python
#Typical Parameters for the atmosphere:

t = 1
mu = 1.6 #Value for greatest instability
U = 10
L = 4000e3 #Length of the channel
y = L/10
l = np.pi/L #n = 1
Ld = 1000e3 #Deformation radius
k = mu/Ld
sigma = 0.3*(U/Ld) #Maximum growth rate
cr = 0.5*U         #Real part of c
ci = -sigma/k      #Imaginary part of c. We can also use:
                   #ci = (U/mu)*np.sqrt((-mu/2 + 1/np.tanh(mu/2))*(mu/2 - np.tanh(mu/2)))
    
c = complex(cr,ci)
```

Now we will plot the streamfunction $\psi'$, the temperature $T = \partial_z \psi'$, and the meridional velocity $v' = \partial_x \psi'$. Since we are working with very large parameters, it is convenient to adequate the vertical domain for each function.

```python
#Streamfunction
#Define domain for plot:

x = np.linspace(0, 8e9)
z = np.linspace(0, 40)
X, Z = np.meshgrid(x, z)
P1 = psi(X,Z)

plt.contourf(X, Z, P1, 15,cmap='jet')
plt.colorbar()
plt.title('Streamfunction')
plt.xlabel('x (km)')
plt.ylabel('z (km)')
```

```python
#Temperature
#Define domain for plot:

x = np.linspace(0, 8e9)
z = np.linspace(0, 1e3)
X, Z = np.meshgrid(x, z)
P2 = temperature(X,Z)

plt.contourf(X, Z, P2,15,cmap='jet')
plt.colorbar()
plt.title('Temperature')
plt.xlabel('x (km)')
plt.ylabel('z (km)')
```

```python
#Meridional velocity
#Define domain for plot:

x = np.linspace(0, 8e9)
z = np.linspace(0, 40)
X, Z = np.meshgrid(x, z)
P3 = meridional_velocity(X,Z)

plt.contourf(X, Z, P3, 15,cmap='jet')
plt.colorbar()
plt.title('Meridional velocity')
plt.xlabel('x (km)')
plt.ylabel('z (km)')
```

### Case 2: Stable Waves

```python
#Typical Parameters for the atmosphere:

t = 1
mu = 16 #Greater than the critical value
U = 10
L = 4000e3
y = L/10
l = np.pi/L
Ld = 1000e3
k = mu/Ld
c = (U/2) - (U/mu)*np.sqrt(((mu/2) - 1/np.tanh(mu/2))*((mu/2) - np.tanh(mu/2))) #Note that c is real in this case
```

```python
#Streamfunction
#Define domain for plot:

x = np.linspace(0, 7e8)
z = np.linspace(0,57)
X, Z = np.meshgrid(x, z)
P1 = psi(X,Z)

plt.contourf(X, Z, P1,12,cmap='jet')
plt.colorbar()
plt.title('Streamfunction')
plt.xlabel('x (km)')
plt.ylabel('z (km)')
```
